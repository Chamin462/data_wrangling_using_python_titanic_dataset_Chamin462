# Method of Normalization
 1- Simple feature scaling\
      - x(new)=x(old)/x(max)\
 2- Min-Max Method\
 3- Z-score (standard score) -3 to +3\
 4- Long transformation\

````python
df4.head()
````

````python
df_4 =df4[['age','fare']]
df_4.head()
````

````python
# simple feature scalling
df4['fare']=df4['fare']/df4['fare'].max()
df4['age']=df4['age']/df4['age'].max()
df4.head()
````

````python
# Min Max Method
df4['fare']=(df4['fare']-df4['fare'].min())/(df4['fare'].max()-df4['fare'].min())
df4['age']=(df4['age']-df4['age'].min())/(df4['age'].max()-df4['age'].min())
df4.head()
````

````python
# Z-score (standard score)
df4['fare']=(df4['fare']-df4['fare'].mean())/df4['fare'].std()
df4['age']=(df4['age']-df4['age'].mean())/df4['age'].std()
df4.head()
````

````python
# log transformation
df4['fare']=np.log(df4['fare'])
df4.head()
````

# Binning Data in Pandas with cut and qcut

# What is Binning in Pandas and Python?
In many cases when dealing with continuous numeric data (such as ages, sales, or incomes), it can be helpful to create bins of your data. Binning data will convert data into discrete buckets, allowing you to gain insight into your data in logical ways. Binning data is also often referred to under several other terms, such as discrete binning, quantization, and discretization.

In this tutorial, youâ€™ll learn about two different Pandas methods, .cut() and .qcut() for binning your data. These methods will allow you to bin data into custom-sized bins and equally-sized bins, respectively. Equal-sized bins allow you to gain easy insight into the distribution, while grouping data into custom bins can allow you to gain insight into logical categorical groupings.
# Pandas qcut: Binning Data into Equal-Sized Bins
The Pandas .qcut() method splits your data into equal-sized buckets, based on rank or some sample quantiles. This process is known as quantile-based discretization. Letâ€™s take a look at the parameters available in the function:
# Parameters of the Pandas .qcut() method
```
 pd.qcut(
    x,                      # Column to bin
    q,                      # Number of quantiles
    labels=None,            # List of labels to include
    retbins=False,          # Whether to return the bins/labels or not
    precision=3,            # The precision to store and display the bins labels
    duplicates='raise'      # If bin edges are not unique, raise a ValueError
) 
```

````python
# Checking the data type of the qcut column
df5['Age Groups'] = pd.qcut(df5['age'], 4)
print(df5['Age Groups'].dtype)
````

````python
print(df5['Age Groups'].dtype)
````

````python
# Splitting Age Column into Four Quantiles
df5['Age Groups'] = pd.qcut(df5['age'],[0, 0.25, 0.5, 0.75, 1])
df5.head()
````

````python
# Adding Labels to Pandas .qcut()
df5['Age Groups'] = pd.qcut(df5['age'],[0, 0.25, 0.5, 0.75, 1],labels=['0-25%', '26-49%', '51-75%', '76-100%'])
df5.head()
````

````python
# Modifying Precision in Categories
df5['Age Groups'] = pd.qcut(df5['age'],4,precision=1)
df5.head()
````

````python
# Parameters of the .cut() Function
###
# pd.cut(
#   x,                          # The input array to be binned
#  bins,                       # The bins to use: int (# of bins) or sequence (widths) 
  #  right=True,                 # Whether to include right-most edge
 #   labels=None,                # Labels to be used for bins
  #  retbins=False,              # Whether to return bins or not
 #   precision=3,                # Precision to store and display bins
  #  include_lowest=False,       # Whether first interval should be left inclusive or not
  #  duplicates='raise',         # What to do if bins edges are not unique
 #   ordered=True                # Whether labels are ordered or not
#)
````

````python
df5['Age Group'] = pd.cut(df5['age'],[0, 17, 64, 100])
df5.head()
````

````python
# Adding labels to the groupings
df5['Age Group'] = pd.cut(df5['age'],[0, 17, 64, 100], labels=['0-18 years old', '18-65 years old', '65+ years old'])
df5.head()
````

````python
# Using the right= argument to modify binning behavior
df5['Age Group'] = pd.cut(df5['age'],[0, 18, 65, 100],labels=['0-18 years old', '18-65 years old', '65+ years old'],right=False)
df5.head()
````

````python
# Including left-most values
df5['Age Group'] = pd.cut(df5['age'],[0, 18, 65, 100],labels=['0-18 years old', '18-65 years old', '65+ years old'],include_lowest=True)
df5.head()
````

````python
# Creating Ordered Categories
print(pd.cut( df5['age'],[0, 18, 65, 100],labels=['0-18 years old', '18-65 years old', '65+ years old'],ordered=True))
````
